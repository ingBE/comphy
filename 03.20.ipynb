```python
f=open('Romeo header.txt','r')

aa, bb, cc = [], [], []

for ii in f.readlines():
    aa.append(ii)
    
print(aa)
for ii in aa:
    bb.extend(ii.split())

print(bb)

for ii in bb:
    if not(ii in cc):
            cc.append(ii)

print(cc)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6.2,0.2)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'--s',x,z,'-o')
plt.title('Sine and cosine')
plt.ylabel('Y values')
plt.xlabel('X values')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6.2,0.2)
y = np.sin(x)
z = np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y,'-o')
plt.title('Sine')
plt.ylabel('Y values')
plt.xlabel('X values')

plt.subplot(1,2,2)
plt.plot(x,z,'-o')
plt.title('Sine')
plt.ylabel('Y values')
plt.xlabel('X values')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6.2,0.2)
y = np.sin(x)
z = np.cos(x)

plt.subplot(2,2,1)
plt.plot(x,y,'-o')
plt.title('Sine')
plt.ylabel('Y values')
plt.xlabel('X values')

plt.subplot(2,2,2)
plt.plot(x,z,'-o')
plt.title('Sine')
plt.ylabel('Y values')
plt.xlabel('X values')

plt.subplot(2,2,3)
plt.plot(x,exp(-x**2.0),'-o')
plt.title('Gaussian')
plt.ylabel('Y values')
plt.xlabel('X values')

plt.subplot(2,2,4)
plt.plot(x,np.log(x),'-o')
plt.title('Log')
plt.ylabel('Y values')
plt.xlabel('X values')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6.2,0.2)
y = [np.sin(x), np.cos(x), np.exp(-x**2.0), np.log(x)]
name = ['Sine', 'Cosine', 'Gaussian', 'Log']

for i in range(0,4):
    plt.subplot(2,2,i+1)
    plt.plot(x,y[i],'-o')
    plt.title(name[i])

plt.show()
```

```python
#problem
#Create three subplots, 
#and plot respectively y=x, y=x^2 and y=x^3
#for x ranging between 0 to 10.
#Distinguish the graphs by different symbols.

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,10,0.2)
y = [x, x**2, x**3]
name = ['x', 'x^2','x^3']

for i in range(0,3):
    plt.subplot(3,1,i+1)
    plt.plot(x,y[i],'-o')
    plt.title(name[i])
    
plt.show()
```
